```python
import math


def calcDigits(num):
    
    numOfDgts=0
   
    while(num!=0):
        num=num/10
        num=math.floor(num)
        numOfDgts+=1
    return numOfDgts

def reverse(num):
    arr=str(num)
    rev=int(arr[::-1])        
    return rev

def checkPalindrome(num):
    count=calcDigits(num)#total number of digits
    hcount=int(count/2)#number of iterations for loop
    check=0#variable to check wheter number is palindrom or not
    arr=str(num)

    for i in range(hcount):
        if arr[i]==arr[count-i-1]:
            check+=1

    if check==hcount:
        return True
    else:
        return False

def sumIsPalindromeOrNot(num):
    for i in range(50):
        rev=reverse(num)
        sum=num+rev
        num=sum
        if checkPalindrome(sum)==True:
            return True
        
        if i==49:
            return False

count=0
limit=10000
for i in range(0,limit):
    check=sumIsPalindromeOrNot(i)

    if check==True:
        count+=1

print("Lychrel numbers Below",limit,":",limit-count)



#num=123
#rev=reverse(num)
#print(num+rev)
```
